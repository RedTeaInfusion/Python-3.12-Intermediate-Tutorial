# IMPORT

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# DATA

```python
data = {
    'Month': ['JAN', 'FEB', 'MAR', 'APR', 'MAY'],
    'Sales': [250, 300, 400, 500, 200],
    'Expenses': [100, 110, 200, 300, 150]
}
df = pd.DataFrame(data)
print('Dataframe:')
print(df)
```

# PROFIT

```python
df['Profit'] = df['Sales'] - df['Expenses']
print('Dataframe with profit column')
print(df)
```

# SUMMARY

```python
print('Summary statistics:')
print(df.describe())
```

# GRAPH

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Sales'], label='Sales', marker='o')
plt.plot(df['Month'], df['Expenses'], label='Expenses', marker='o')
plt.plot(df['Month'], df['Profit'], label='Profit', marker='o')

plt.title('Monthly data')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
